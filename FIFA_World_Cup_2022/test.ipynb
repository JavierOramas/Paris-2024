```python
import seaborn as sns
import pandas as pd
import json
```

```python
df = pd.read_csv('results.csv')
with open('groups.json') as f:
    groups = json.load(f)
df
```

```python
import matplotlib.pyplot as plt
labels = [i for i in df["Team"]]
fig, ax = plt.subplots(figsize=(10,10))
# sort dataframe by 4th finals
df = df.sort_values(by=['Winner'], ascending=False)
ax = sns.heatmap(data=df[["4th Finals", "Semi Finals", "Final", "Winner"]],vmin=0,vmax=1,linewidths=2, linewidth='.5',yticklabels=labels, annot=True, cmap="crest")
plt.yticks(rotation=0)
# plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom=False, top = False, labeltop=True)
ax.xaxis.tick_top() # x axis on top
ax.xaxis.set_label_position('top')
plt.plot()
plt.savefig(f"results.jpg", dpi=1000 )
plt.close()
```

```python
import matplotlib.pyplot as plt
for g in groups.keys():
    df1 = df[df['Team'].isin(groups[g])]
    labels = [i for i in df1["Team"]]
    fig, ax = plt.subplots(figsize=(8,8))
    ax = sns.heatmap(data=df1[["1st Group","2nd Group"]],vmin=0,vmax=1,linewidths=2, linewidth='.5',yticklabels=labels, annot=True, cmap="crest")
    plt.savefig(f"{g}.jpg", dpi=1000 )
    plt.close()
```
